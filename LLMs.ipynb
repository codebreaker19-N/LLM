```python
from transformers import AutoModelForCausalLM, AutoTokenizer

# Load model and tokenizer
model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3-mini-4k-instruct",
    device_map="auto",
    torch_dtype="auto",
    trust_remote_code=False,
)
tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-4k-instruct")
```

**AutoModelForCasualLM** --
 “Auto” → automatically selects the correct model architecture
“CausalLM” → Causal Language Model
Used for text generation
Predicts the next word based on previous words

**AutoTokenizer **-- converts texts to numbers
**"microsoft/Phi-3-mini-4k-instruct" **
   Phi-3 → model family
   mini → smaller, faster version
   4k → supports ~4000 tokens context
   instruct → trained to follow instructions (chat-style)
**device_map="auto"**
   Automatically decides where to load the model
**torch_dtype = "auto"**
 Automatically choose the best type based on your hardware
** trust_remote_code=False**
Prevents execution of custom code from the model repository

```python
from transformers import pipeline

# Create a pipeline
generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    return_full_text=False,
    max_new_tokens=500,
    do_sample=False
)
```

**from transformers import pipeline**
pipeline comes from the Transformers
It provides a high-level API to run models in just a few lines
**generator** text generation system
